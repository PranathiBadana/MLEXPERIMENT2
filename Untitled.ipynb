```python
import pandas as pd

columns = [
    "SepalLength",
    "SepalWidth",
    "PetalLength",
    "PetalWidth",
    "Species"
]

df = pd.read_csv("iris.data", names=columns)

print(df.head())
```

```python
print(df.head())
```

```python
print(df.info())
```

```python
print(df.describe())
```

```python
data = df["PetalLength"]

print("Mean =", data.mean())
print("Median =", data.median())
print("Mode =", data.mode()[0])
print("Variance =", data.var())
print("Standard Deviation =", data.std())
```

```python
q1 = data.quantile(0.25)
q2 = data.quantile(0.50)
q3 = data.quantile(0.75)

print("Q1 =", q1)
print("Q2 =", q2)
print("Q3 =", q3)

iqr = q3 - q1
print("IQR =", iqr)
```

```python
data = df["PetalLength"]

print("Mean =", data.mean())
print("Median =", data.median())
print("Mode =", data.mode()[0])
print("Variance =", data.var())
print("Standard Deviation =", data.std())
```

```python
q1 = data.quantile(0.25)
q2 = data.quantile(0.50)
q3 = data.quantile(0.75)

print("Q1 =", q1)
print("Q2 =", q2)
print("Q3 =", q3)

iqr = q3 - q1
print("IQR =", iqr)
```

```python
lower = q1 - 1.5 * iqr
upper = q3 + 1.5 * iqr

outliers = data[(data < lower) | (data > upper)]

print("Lower Limit =", lower)
print("Upper Limit =", upper)
print("Outliers:")
print(outliers)
```

```python
import matplotlib.pyplot as plt

plt.hist(data)
plt.title("Histogram of Petal Length")
plt.xlabel("Petal Length")
plt.ylabel("Frequency")
plt.show()
```

```python
plt.boxplot(data)
plt.title("Boxplot of Petal Length")
plt.show()
```
